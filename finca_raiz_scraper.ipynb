```python
import time
import requests as rq
from bs4 import BeautifulSoup

from selenium import webdriver
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import StaleElementReferenceException
from selenium.common.exceptions import TimeoutException

from selenium.webdriver.common.by import By
```

```python
# Getting the url
url = 'https://www.fincaraiz.com.co/apartaestudios-apartamentos-casas/proyectos-vivienda-nueva?ubicacion=Colombia'
# Creating the request
r = rq.get(url)
# Verifying if the request is accepted
r.status_code
```

```python
# Instantiating the driver
print('Instanciando el driver')
options = webdriver.ChromeOptions()
options.add_argument('--incognit')
driver = webdriver.Chrome(executable_path = r"C:\Users\jeanf\OneDrive\Desktop\Estudios\Data Science - Platzi\Curso de Web Scrapping extracción de datos en la web\SELENIUM SCRAPER\chromedriver_win32\chromedriver.exe", options = options)
# Opening URL
driver.get(url)
driver.maximize_window()
time.sleep(3)
```

```python
def get_price(driver,apartments):
    apartment = apartments[0]

    price_list = []
   
    for apartment in apartments:
        try:
            # price_wait = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, './/span[@class = "MuiTypography-root jss349 MuiTypography-body1"]'))) -> This didn't work for me
            price_list.append(((apartment.find_element(By.XPATH, './/span[@class = "MuiTypography-root jss349 MuiTypography-body1"]').text)))
        except NoSuchElementException:
            price_list.append('None')
            print('IMPORTANTE: *Algunos elementos faltantes*')
    return price_list

def get_location(driver,apartments):
    apartment = apartments[0]
    location_list = []
   
    for apartment in apartments:
        try:
            location_list.append(apartment.find_element(By.XPATH, './/div[@class = "MuiGrid-root MuiGrid-container MuiGrid-item MuiGrid-grid-xs-12"]').text)
        except NoSuchElementException:
            location_list.append('None')
            print('IMPORTANTE: *Algunos elementos faltantes*')
    return location_list

def get_size(driver,apartments):
    apartment = apartments[0]
    size_list = []
   
    for apartment in apartments:
        try:
            size_list.append(apartment.find_element(By.XPATH, './/div[@class = "MuiGrid-root jss352 MuiGrid-container MuiGrid-item MuiGrid-grid-xs-12"]//span[1]').text)
        except NoSuchElementException:
            size_list.append('None')
            print('IMPORTANTE: *Algunos elementos faltantes*')
    return size_list

def get_rooms(driver,apartments):
    apartment = apartments[0]
    rooms_list = []
   
    for apartment in apartments:
        try:
            rooms_list.append(apartment.find_element(By.XPATH, './/div[@class = "MuiGrid-root jss352 MuiGrid-container MuiGrid-item MuiGrid-grid-xs-12"]//span[3]').text)
        except NoSuchElementException:
            rooms_list.append('None')
            print('IMPORTANTE: *Algunos elementos faltantes*')
    return rooms_list

def get_baths(driver,apartments):
    apartment = apartments[0]
    baths_list = []
   
    for apartment in apartments:
        try:
            baths_list.append(apartment.find_element(By.XPATH, './/div[@class = "MuiGrid-root jss352 MuiGrid-container MuiGrid-item MuiGrid-grid-xs-12"]//span[5]').text)
        except NoSuchElementException:
            baths_list.append('None')
            print('IMPORTANTE: *Algunos elementos faltantes*')
    return baths_list

def get_links(driver,apartments):
    apartment = apartments[0]
    links_list = []
   
    for apartment in apartments:
        try:
            links_list.append(apartment.find_element(By.XPATH, './/a[@class = "MuiTypography-root MuiLink-root MuiLink-underlineHover jss362 jss341 MuiTypography-colorPrimary"]').get_attribute('href'))
        except NoSuchElementException:
            links_list.append('None')
            print('IMPORTANTE: *Algunos elementos faltantes*')
    return links_list

def scrapper():
    price_list = []
    location_list = []
    size_list = []
    rooms_list = []
    baths_list = []
    links_list = []

    for page in range(1,41):
        # instantiating the driver
        options = webdriver.ChromeOptions()
        options.add_argument('--incognit')
        driver = webdriver.Chrome(executable_path = r"C:\Users\jeanf\OneDrive\Desktop\Estudios\Data Science - Platzi\Curso de Web Scrapping extracción de datos en la web\SELENIUM SCRAPER\chromedriver_win32\chromedriver.exe", options = options)
        print('Instanciando el driver para página {}'.format(page))
        time.sleep(3)
        print('Abriendo la URL...')
        # Opening the url
        url = 'https://www.fincaraiz.com.co/apartaestudios-apartamentos-casas/proyectos-vivienda-nueva?ubicacion=Colombia' + '&pagina=' + str(page)
        driver.get(url)
        driver.maximize_window()
        time.sleep(5)
        # Finding the class where we can have all the apartments
        apartments = driver.find_elements(By.XPATH, './/div[@class = "MuiGrid-root MuiGrid-item MuiGrid-grid-xs-12 MuiGrid-grid-sm-6 MuiGrid-grid-lg-4 MuiGrid-grid-xl-4"]')
        print('Se encontraron: {} apartamentos en la página'.format(len(apartments)))
        print('Extrayendo la información...')
        # Prices
        price_list += (get_price(driver, apartments))
        
        print('Precios obtenidos...')
        # Location
        location_list += get_location(driver, apartments)
        
        print('Ubicaciones obtenidas...')
        # Size
        size_list += get_size(driver, apartments)
        
        print('Tamaños obtenidos...')
        # Rooms
        rooms_list += get_rooms(driver, apartments)
        
        print('Rooms obtenidos...')
        # Bathrooms
        baths_list += get_baths(driver, apartments)
        
        print('Baños obtenidos...')
        # Link
        links_list += get_links(driver, apartments)
        
        print('Links obtenidos...')
        
        data_dict = {
            'Location': location_list,
            'Price': price_list,
            'Size': size_list,
            'Rooms': rooms_list,
            'Bathrooms': baths_list,
            'link': links_list
            
        }

        print('Información scrapeada \n')
        driver.close()

    print('Scraper Finalizado\n')
    return data_dict

data_dict = scrapper()
data_dict
```

```python
import pandas as pd
# And finally creating the DataFrame using pandas!
df = pd.DataFrame(data_dict)
df
```

```python
# And exporting the csv file...
df.to_csv('finca_raiz_apartments.csv')
```
